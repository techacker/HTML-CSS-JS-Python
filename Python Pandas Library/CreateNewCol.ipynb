# Creating New Columns derived from existing columns

```python
import pandas as pd
```

```python
air_quality = pd.read_csv('Data/air_quality_no2.csv', index_col=0, parse_dates=True)
air_quality.head()
```

```python
# Express NO2 concentration of the station in London in mg/m3
# Assuming a conversion factor of 1.882

air_quality['london_mg_per_cubic'] = air_quality['station_london'] * 1.882
air_quality.head()
```

```python
# Ratio of the values in Paris vs Antwerp and save the result in a new col
air_quality['ratio_paris_antwerp'] = (air_quality['station_paris'] / air_quality['station_antwerp'])
air_quality.head()
```

```python
# Rename columns
air_quality_renamed = air_quality.rename(columns={
    'station_antwerp': 'BETR801',
    'station_paris': 'FR04014',
    'station_london': 'London Westminster'
})
# The rename() function can be used for both rows and columns labels. Provide a dictionary with the keys.
```
